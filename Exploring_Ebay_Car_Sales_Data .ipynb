The NaN columns are immediate candidates to be dropped from the dataframe, because it isn't clear what that data represents. Also, there are the two columns "price" and "odometer" that contains numerical data as string dtypes. 

```python
autos["price"] = (autos["price"]
                  .str.replace("$","")
                  .str.replace(",","")
                  .astype(int))
autos["odometer"] = (autos["odometer"]
                     .str.replace(",","")
                     .str.replace("km","")
                     .astype(int))
autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
autos.head()
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index

common_brands
```

```python
av_brand_prices = {}
for brand in common_brands:
    brand_name = autos.loc[autos["brand"] == brand]
    brand_ave_price = brand_name["price"].mean()
    av_brand_prices[brand] = int(brand_ave_price)

av_brand_prices
```

```python
average_mileage = {}
for brand in common_brands:
    brands = autos.loc[autos["brand"] == brand]
    brand_ave_mileage = brands["odometer_km"].mean()
    average_mileage[brand] = int(brand_ave_mileage)
    
average_mileage
```

```python
av_mileage = pd.Series(average_mileage).sort_values(ascending=False)
av_prices = pd.Series(av_brand_prices).sort_values(ascending=False)
av_prices
```

```python
top_brands = pd.DataFrame(av_mileage, columns = ["average_mileage"])
top_brands
```

```python
top_brands["average_prices"] = av_prices
top_brands
```
